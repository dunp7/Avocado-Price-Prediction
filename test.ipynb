```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import datetime as dt
from sklearn.linear_model import LinearRegression
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from Function import *
```

```python
data = pd.read_csv('Avocado.csv', index_col=0)
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
data1 =data[(data['year']==2015) & (data['region']=='Orlando') & (data['type']=='organic')].sort_values(by= 'Date')
data2 =data[(data['year']==2015) & (data['region']=='Orlando') & (data['type']=='conventional')].sort_values(by= 'Date')
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data1['Date'], data1['AveragePrice'], marker='o', linestyle='-')
plt.plot(data2['Date'], data2['AveragePrice'], marker='o', linestyle='-')

plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
window_size = 4 
ma_smoothed = data1['AveragePrice'].rolling(window=window_size).mean()

# Plot the original time series
plt.figure(figsize=(20, 10))
plt.plot(data1['Date'], ma_smoothed, marker = 'o' ,color='red', label='MA Smoothed')
plt.title('Time Series Plot with MA Smoothing')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
data1['Value_diff'] = data1['AveragePrice'].diff()
data2['Value_diff'] = data2['AveragePrice'].diff()

# Plot the original time series
plt.figure(figsize=(20, 10))
plt.plot(data1['Date'], data1['Value_diff'], marker = 'o' ,color='red', label='MA Smoothed')
plt.plot(data2['Date'], data2['Value_diff'], marker = 'o' ,color='blue', label='MA Smoothed')

plt.title('Time Series Plot with differcing')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
