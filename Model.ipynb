```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error



```

```python
data = pd.read_csv('train.csv', index_col= 0)
data = data[['Date','AveragePrice','Value_diff_1','Value_diff_2']]
data
```

```python
# Split data into training and testing sets
train = data[:500]
test = data[500:]

# Fit ARMA model
arma_model = ARIMA(train, order=(2, 0, 2))  # Example order, replace with appropriate values
arma_result = arma_model.fit()

# Fit SARIMA model
sarima_model = SARIMAX(train, order=(1, 0, 1), seasonal_order=(0, 0, 0, 12))  # Example order, replace with appropriate values
sarima_result = sarima_model.fit()

# Predictions
arma_pred = arma_result.predict(start=test.index[0], end=test.index[-1])
sarima_pred = sarima_result.predict(start=test.index[0], end=test.index[-1])

# Evaluation metrics
arma_mse = mean_squared_error(test, arma_pred)
sarima_mse = mean_squared_error(test, sarima_pred)

arma_mae = mean_absolute_error(test, arma_pred)
sarima_mae = mean_absolute_error(test, sarima_pred)

print("ARMA - MSE:", arma_mse)
print("SARIMA - MSE:", sarima_mse)

print("ARMA - MAE:", arma_mae)
print("SARIMA - MAE:", sarima_mae)

# Optionally, visualize predictions
plt.figure(figsize=(10, 6))
plt.plot(test, label='Actual')
plt.plot(arma_pred, label='ARMA Predictions')
plt.plot(sarima_pred, label='SARIMA Predictions')
plt.legend()
plt.show()
```
